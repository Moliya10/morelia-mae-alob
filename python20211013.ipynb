# leap year

```python
year= eval(input("Please input the year"))

if ((year%400==0) or (year%4==0 and year%100!=0)):
  print("{0} is a leap year".format(year))
else:
  print("{0} is not a leap year".format(year))
```

## For loop

```python
Desserts = ["ice cream", "banana split", "cake"]
for x in Desserts:
  print(x)
```

```python
for x in "banana split":
  print(x)
```

```python
for x in range(11):
  print(x)
```

```python
fruits = ["watermelon", "grapes", "mangoes"]
for x in fruits:
  print(x)
  if x == "watermelon":
    break
```

```python
for x in range(3):
  print(x)
  if x == 2:
    break
```

```python
for x in range(13):
    if x == 9:
        continue
    print(x)
```

```python
for x in range(10):
  if x%5 == 0:
   continue
  
  print(x)
```
